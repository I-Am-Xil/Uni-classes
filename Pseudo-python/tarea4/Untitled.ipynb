```python
import matplotlib.pyplot as plt
import csv

x = []
y = []

with open("ejemplo.csv", 'r') as csvfile:
    plots = csv.reader(csvfile, delimiter = ',')
    for row in plots:
        x.append(int(row[0]))
        y.append(int(row[1]))
        
plt.plot(x, y, label = "Datos interesantes!")
plt.xlabel('x')
plt.ylabel('y')
plt.title("Grafica interesante\ncon dos picos")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x, y = np.loadtxt("ejemplo.csv", delimiter = ',', unpack = True)

plt.plot(x, y, label = "Datos interesantes!")
plt.xlabel('x')
plt.ylabel('y')
plt.title("Grafica interesante\ncon dos picos")
plt.legend()

#plt.savefig("picos.pdf") Guardar como pdf
#plt.savefig("picos.png") Guardar como png

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x, y = np.loadtxt("ejemplo.csv", delimiter = ',', unpack = True)

plt.plot(x, y, 'o', label = "Datos interesantes!")
plt.plot(x, y, 'ro-', label = "Datos interesantes!")
plt.xlabel('x')
plt.ylabel('y')
plt.title("Grafica interesante\ncon dos picos")
plt.legend()

#plt.savefig("picos.pdf") Guardar como pdf
#plt.savefig("picos.png") Guardar como png

plt.show()
```
